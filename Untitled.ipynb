#### FOCUSSING ON GENRES COLUMNS

```python
movies.iloc[0].genres #the data present in it is not in readable manner ie list of dictionary
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# we have to convert this in the following manner: ['Action''Adventure','Fantasy','SciFi']
```

```python
#CREATING A HELPER FUNCTION:
#the above one is in string format so we will use ast module to convert it into list and then we will perform our loop
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert) #for every movie i am getting the genres of that movie
```

```python
movies['genres'] = movies['genres'].apply(convert)#storing in the same column
```

```python
movies.head() #now if we see the genres column then its more clarified now compared it with before
```

#### Doing the same thing with our keywords column:

```python
movies['keywords'].apply(convert) # we will gte the list of all the keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert) #keepimg it in variable
```

```python
movies.head() #now the keywords in more clarified
```

#### Now we will do the same thing with our cast column:

```python
movies['cast'][0] 
```

```python
# we have to only get the first 3 actors list and we have to go in the first 3 dictionary and pull out the name column 
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head() #now we have solved the cast problem
```

#### Solving crew now:its difficult cause there are many dictionaries and we have to find only that where job value is of 'director'

```python
# creating a function
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director': #if the job value is director
            L.append(i['name']) #then only we will append if its director
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head() #now the we can see director in crew column
```

#### Problem: The space present in the words of column genres,keywords,cast.crew ..we have to remove it and use TRANSFORMATION, EG:Sam Worthington=SamWorthington

#### It is very important cause it will remove the confusion in our model .EG: Sam Worthington(from cast) AND Sam Mendes(from crew), our model will get confused so to remove it we have to remove spaces

```python
#APPLYING THE TRANSFORMATION TO THE COLUMNS:

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")  for i in x]) #The space has been removed from Science fiction
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")  for i in x]) 
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")  for i in x]) 
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")  for i in x]) 
```

### STEMMING USING nltk:

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#EXAMPLE:
ps.stem('danced')
```

```python
# Making new function:
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y) #converting this list into string
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## VECTORIZATION OF DATA:

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
#The CountVectorizer gives the data in the form of sci-py spars matrix so we will use numpy to convert it into numpy array

cv.fit_transform(new_df['tags']).toarray()
```

```python

cv.fit_transform(new_df['tags']).toarray().shape #we have 4806 movies
```

```python
#naming it into vectors variable
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names() #we solved the problem using stemming and words like actors and actor are not repeating
```

### USING THE COSINE DISTANCE FOR THE MEASUREMENT OF THE SIMILARITIES BETWEEN THE MOVIES:

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape # The total distance we are calculating is 4806 multiplied by 4806
```

```python
similarity = cosine_similarity(vectors)
```

#### it is showing the similarity score of first movie with each and every movie that is with all 4806 movies,the closer to one the more similar they are 

```python
similarity[0]
```

## *Now we creating the function which will give us the 5 recommended movies which the user wants

```python
# using enumerate function: we are getting the list of tuples and we are not losing the index position 
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list: 
        print(i[0])
```

```python
recommend('Avatar') #its giving the index number
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list: 
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar') #now we are getting the movies names
```
